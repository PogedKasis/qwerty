```python
import matplotlib.pyplot as plt
%matplotlib inline

def to_month_year(x):
    return (pd.to_datetime(x).year, pd.to_datetime(x).month)

def len_list(x):
    return len(x)

def str_date(x):
    return str(x[1]) + '.' + str(x[0])


filter_hw = (df['is_homework'] == True) & ((df['status'] == 'start') | (df['status'] == 'done') | (df['status'] == 'fail'))
hw_start = pd.DataFrame((df.loc[filter_hw].groupby(['title', 'progress_id'])['start_date']).agg('min'))
metric_df = hw_start.copy()   ### Датафрейм для метрики среднего времени выполнения дз по курсам (к заданию 5)
hw_start['start_date'] = hw_start['start_date'].apply(to_month_year)

students_number_df = pd.DataFrame(hw_start.groupby(['title', 'start_date'])['start_date'].agg(len_list)) # <-- датафрейм с приростом
students_number_df.rename(columns={'start_date': 'students_number'}, inplace=True)                       # по курсам-месяцам

period = pd.date_range(start ='3-1-2016', end ='8-1-2019', freq ='1M').to_list()
X = list(map(lambda x: (x.year, x.month), period))
X = pd.DataFrame({'date': X})
print('Прирост студентов на каждом курсе в каждом месяце за всю историю: \n')
for i in df['title'].unique():
    title_i = students_number_df.loc[i]
    x_y = pd.merge(X, title_i, how='left', left_on='date', right_index=True).fillna(0)
    x_y['date'] = x_y['date'].apply(str_date)
    plt.rcParams['figure.figsize'] = [10, 5]
    plt.plot(x_y['date'], x_y['students_number'], '.', linestyle='-')
    plt.xticks(rotation='vertical')
    plt.title(i)
    plt.grid()
    plt.show()
```

```python
def to_year_month(x):
    x = pd.to_datetime(x)
    x = (x.year, x.month)
    return x


df['start_date'] = df['start_date'].apply(to_year_month)
```

```python
def delta_func(x):
    start = x[0]
    finish = x[1]
    if pd.isnull(finish[0]):
        finish = (2019, 7)
        
    years = list(range(start[0], finish[0] + 1))
    delta = []
    if len(years) == 1:
        for i in range(start[1], finish[1] + 1):
            delta.append((start[0], i))
    else:
        for i in range(start[1], 13):
            delta.append((start[0], i))
            
        for i in range(years[1], years[-1]):
            for j in range(1, 13):
                delta.append((i, j))
                
        for i in range(1, finish[1]):
            delta.append((finish[0], i))
            
    return delta

hw_df['delta_hw'] = list(map(lambda x: delta_func(x), hw_df[['start_date', 'finish_date']].values))
```
